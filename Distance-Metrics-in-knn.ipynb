#### A practical rule for KNN

When choosing a distance metric, ask yourself: What kind of features do I have?

Numerical features (height, weight, age, income): Start with Euclidean distance, but standardize or normalize the features first. Otherwise, variables with larger scales will dominate the distance.

Binary or categorical features: Consider Hamming distance.

Text or embedding vectors: Use Cosine similarity/distance because orientation matters more than magnitude.

If you suspect outliers or data with many dimensions: Try Manhattan distance.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([2, 3])
B = np.array([7, 8])

print("Point A:", A)
print("Point B:", B)
```

```python
plt.figure(figsize=(6,6))

plt.scatter(A[0], A[1], s=100, label="Point A")
plt.scatter(B[0], B[1], s=100, label="Point B")

plt.plot([A[0], B[0]], [A[1], B[1]], '--')

plt.xlim(0,10)
plt.ylim(0,10)

plt.grid(True)
plt.legend()

plt.title("Two Points")

plt.show()
```

```python
distance = np.sqrt(np.sum((A-B)**2))

print("Euclidean Distance =", distance)
```

```python
plt.figure(figsize=(6,6))

plt.scatter(A[0], A[1], s=120)
plt.scatter(B[0], B[1], s=120)

plt.plot([A[0], B[0]], [A[1], B[1]], linewidth=3)

plt.text(4.2,5.3,"Shortest Distance")

plt.grid(True)

plt.title("Euclidean Distance")

plt.show()
```

```python
distance = np.sum(np.abs(A-B))

print("Manhattan Distance =", distance)
```

```python
plt.figure(figsize=(6,6))

plt.scatter(A[0], A[1], s=120)
plt.scatter(B[0], B[1], s=120)

plt.plot([A[0],B[0]],[A[1],A[1]],linewidth=3)
plt.plot([B[0],B[0]],[A[1],B[1]],linewidth=3)

plt.grid(True)

plt.title("Manhattan Distance")

plt.show()
```

```python
euclidean = np.sqrt(np.sum((A-B)**2))
manhattan = np.sum(np.abs(A-B))

print("Euclidean :",euclidean)
print("Manhattan :",manhattan)
```

```python
def minkowski(A,B,p):
    return np.sum(np.abs(A-B)**p)**(1/p)

print("p=1 :",minkowski(A,B,1))
print("p=2 :",minkowski(A,B,2))
print("p=3 :",minkowski(A,B,3))
print("p=5 :",minkowski(A,B,5))
```

```python
A = np.array([1,0,1,1,0,1])
B = np.array([1,1,1,0,0,1])

distance = np.sum(A!=B)

print("Hamming Distance =",distance)
```

```python
A = np.array([2,3])
B = np.array([7,8])

similarity = np.dot(A,B)/(np.linalg.norm(A)*np.linalg.norm(B))

print("Cosine Similarity =",similarity)
```

```python
print("Euclidean :",np.sqrt(np.sum((A-B)**2)))
print("Manhattan :",np.sum(np.abs(A-B)))
print("Minkowski (p=3):",minkowski(A,B,3))
print("Cosine Similarity :",similarity)
```
