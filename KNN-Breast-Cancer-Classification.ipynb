### Feature Scaling (Normalization/Standardization)

We fit the scaler **only on the training data** to prevent **data leakage**. If the scaler is fitted on the test data as well, it learns information (like the mean and standard deviation) from data that is supposed to be unseen, making the evaluation unfair and overly optimistic.


fit() learns the mean and standard deviation of each feature from the training data. transform() uses those learned values to scale features using (x - mean) / std. We transform both training and test data so features are on the same scale, but we fit only on the training data to avoid data leakage, since fitting on the test data would use information that should remain unseen.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
# can also be in following
# scaler.fit(x_train)
# X_train = scaler.transform(x_train)
X_test = scaler.transform(x_test)
```

### KNN Predictions

Suppose WE have one test patient.

KNN does the following:

Takes that one test sample.

Computes its distance to every training sample.

Gets one distance value per training sample.

Sorts the distances from smallest to largest.

Selects the K nearest neighbors.

Looks at their labels.

Predicts the class by majority vote.

NumPy Broadcasting Rules: Compare the shapes of the two arrays from right to left. If one array has fewer dimensions, prepend 1s to its shape until both have the same number of dimensions. Then compare each pair of dimensions. A pair is compatible if they are equal or at least one of them is 1. If every pair is compatible, broadcasting succeeds, and the resulting shape takes the maximum value of each dimension (e.g., 1 and 5 become 5). If any pair is neither equal nor 1, broadcasting fails with an error.

```python
one_sample = X_test[0]

diff = one_sample - X_train
dist = np.sqrt(np.sum(diff**2, axis=1))
# axis = 1: Sum across the columns of each row. (row-wise)
```

```python
distances_sorted = np.argsort(dist)
# sort by indices
```

```python
k = 1
nearest_indices = distances_sorted[:k]
nearest_labels = y_train[nearest_indices]
nearest_labels
```

```python
prediction = np.bincount(nearest_labels).argmax()
# np.bincount(nearest_labels) counts how many times each class appears
# if [1,1,1] then predicts [0,3] means class 0 - 0 times and 1 - 3 times
# .argmax() returns the index with the highest count, which is the predicted class.
```

### Full prediction loop for entire test set.

```python
y_pred = []
k = 1
for i in range(X_test.shape[0]):
    one_sample = X_test[i]
    
    diff = one_sample - X_train
    dist = np.sqrt(np.sum(diff**2, axis=1))
        
    nearest_indices = np.argsort(dist)[:k]
    nearest_labels = y_train[nearest_indices]
    
    prediction = np.bincount(nearest_labels).argmax()
    y_pred.append(prediction)

y_pred = np.array(y_pred)
```

```python
accuracy = np.mean(y_pred == y_test)
accuracy
```

```python
print("Correct predictions:", np.sum(y_pred == y_test))
print("Total test samples:", len(y_test))
```

### Voronoi KNN visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.spatial import Voronoi, voronoi_plot_2d

# ---- Reduce to 2D ----
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_train)

# ---- Voronoi based on training points ----
vor = Voronoi(X_2d)

# ---- Plot Voronoi diagram ----
fig, ax = plt.subplots(figsize=(8, 6))
voronoi_plot_2d(vor, ax=ax, show_vertices=False, line_colors='black')

# ---- Color points by class ----
scatter = ax.scatter(X_2d[:, 0], X_2d[:, 1], c=y_train, cmap='coolwarm', s=30)

plt.title("Voronoi Regions (KNN intuition for k=1)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score
def euclidean(x, X_train):
    return np.sqrt(np.sum((X_train - x) ** 2, axis=1))

def manhattan(x, X_train):
    return np.sum(np.abs(X_train - x), axis=1)

def knn_predict(X_train, y_train, x, k, distance_fn):
    dist = distance_fn(x, X_train)
    
    nearest_idx = np.argsort(dist)[:k]
    nearest_labels = y_train[nearest_idx]
    
    return np.bincount(nearest_labels).argmax()

def knn_weighted_predict(X_train, y_train, x, k):
    dist = euclidean(x, X_train)
    
    nearest_idx = np.argsort(dist)[:k]
    nearest_labels = y_train[nearest_idx]
    nearest_dist = dist[nearest_idx]
    
    weights = 1 / (nearest_dist + 1e-8)
    
    vote = {}
    for label, w in zip(nearest_labels, weights):
        vote[label] = vote.get(label, 0) + w
    
    return max(vote, key=vote.get)
```

```python
pred_euc = []

for x in X_test:
    pred = knn_predict(X_train, y_train, x, k=5, distance_fn=euclidean)
    pred_euc.append(pred)

acc_euc = accuracy_score(y_test, pred_euc)
acc_euc
```

```python
pred_man = []

for x in X_test:
    pred = knn_predict(X_train, y_train, x, k=5, distance_fn=manhattan)
    pred_man.append(pred)

acc_man = accuracy_score(y_test, pred_man)
acc_man
```

```python
pred_w = []

for x in X_test:
    pred = knn_weighted_predict(X_train, y_train, x, k=5)
    pred_w.append(pred)

acc_w = accuracy_score(y_test, pred_w)
acc_w
```

```python
print("Euclidean KNN Accuracy:", acc_euc)
print("Manhattan KNN Accuracy:", acc_man)
print("Weighted KNN Accuracy:", acc_w)
```

### K Fold Validation

```python
def create_folds(X, y, n_splits):
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    
    fold_sizes = len(X) // n_splits
    folds = []
    
    for i in range(n_splits):
        start = i * fold_sizes
        end = (i + 1) * fold_sizes
        
        fold_idx = indices[start:end]
        folds.append(fold_idx)
    
    return folds
```

```python
def knn_predict(X_train, y_train, x, k, distance_fn):
    dist = distance_fn(x, X_train)
    nearest_idx = np.argsort(dist)[:k]
    nearest_labels = y_train[nearest_idx]
    return np.bincount(nearest_labels).argmax()
```

```python
def cross_validate_knn(X, y, k_neighbors, n_splits=5):
    
    folds = create_folds(X, y, n_splits)
    accuracies = []
    
    for i in range(n_splits):
        
        # validation fold
        val_idx = folds[i]
        
        # training folds
        train_idx = np.hstack([folds[j] for j in range(n_splits) if j != i])
        
        X_train, y_train = X[train_idx], y[train_idx]
        X_val, y_val = X[val_idx], y[val_idx]
        
        preds = []
        
        for x in X_val:
            pred = knn_predict(X_train, y_train, x, k_neighbors, euclidean)
            preds.append(pred)
        
        acc = np.mean(np.array(preds) == y_val)
        accuracies.append(acc)
    
    return np.mean(accuracies)
```

```python
k_values = [1, 3, 5, 7, 9, 11, 15]

results = {}

for k in k_values:
    acc = cross_validate_knn(X_train, y_train, k_neighbors=k, n_splits=5)
    results[k] = acc
    print("K =", k, "Accuracy =", acc)
```

```python
best_k = max(results, key=results.get)

print("Best K:", best_k)
print("Best CV Accuracy:", results[best_k])
```
