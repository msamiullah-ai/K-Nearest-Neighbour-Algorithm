### Feature Scaling (Normalization/Standardization)

We fit the scaler **only on the training data** to prevent **data leakage**. If the scaler is fitted on the test data as well, it learns information (like the mean and standard deviation) from data that is supposed to be unseen, making the evaluation unfair and overly optimistic.


fit() learns the mean and standard deviation of each feature from the training data. transform() uses those learned values to scale features using (x - mean) / std. We transform both training and test data so features are on the same scale, but we fit only on the training data to avoid data leakage, since fitting on the test data would use information that should remain unseen.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
# can also be in following
# scaler.fit(x_train)
# X_train = scaler.transform(x_train)
X_test = scaler.transform(x_test)
```

```python
## USING NUMPY
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_scaled_train = (x_train - mean)/std
x_scaled_test = (x_test - mean)/std
# we did not compute new mean and std for test data
```

### KNN Predictions

Suppose WE have one test patient.

KNN does the following:

Takes that one test sample.

Computes its distance to every training sample.

Gets one distance value per training sample.

Sorts the distances from smallest to largest.

Selects the K nearest neighbors.

Looks at their labels.

Predicts the class by majority vote.

NumPy Broadcasting Rules: Compare the shapes of the two arrays from right to left. If one array has fewer dimensions, prepend 1s to its shape until both have the same number of dimensions. Then compare each pair of dimensions. A pair is compatible if they are equal or at least one of them is 1. If every pair is compatible, broadcasting succeeds, and the resulting shape takes the maximum value of each dimension (e.g., 1 and 5 become 5). If any pair is neither equal nor 1, broadcasting fails with an error.

```python
one_sample = X_test[0]

diff = one_sample - X_train
dist = np.sqrt(np.sum(diff**2, axis=1))
# axis = 1: Sum across the columns of each row. (row-wise)
```

```python
distances_sorted = np.argsort(dist)
# sort by indices
```

```python
k = 3
nearest_indices = distances_sorted[:k]
nearest_labels = y_train[nearest_indices]
nearest_labels
```

```python
nearest_indices
```

```python
prediction = np.bincount(nearest_labels).argmax()
# np.bincount(nearest_labels) counts how many times each class appears
# if [1,1,1] then predicts [0,3] means class 0 - 0 times and 1 - 3 times
# .argmax() returns the index with the highest count, which is the predicted class.
```
