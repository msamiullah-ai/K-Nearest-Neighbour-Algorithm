```python
from sklearn.datasets import load_wine
import numpy as np
import pandas as pd
data = load_wine()

X = data.data
y = data.target

print(X.shape)
print(y.shape)
print(data.feature_names)
print(data.target_names)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=22
)
```

```python
print(x_train.shape)
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
def knn_predict(X_train, y_train, X_test, k):
    
    # Step 1: Compute Euclidean distance from x_test to every training sample
    distances = np.sqrt(np.sum((X_train - X_test)**2, axis=1))
    
    # Step 2: Find indices of k nearest neighbors
    nearest_indices = np.argsort(distances)[:k]
    
    # Step 3: Get labels of those neighbors
    nearest_labels = y_train[nearest_indices]
    
    # Step 4: Majority voting
    prediction = np.bincount(nearest_labels).argmax()
    
    return prediction
```

```python
y_pred = []

for x in X_test:
    y_pred.append(knn_predict(X_train, y_train, x, k=18))
```

```python
y_pred = np.array(y_pred)
```

```python
accuracy = np.mean(y_pred == y_test)
print("Accuracy:", accuracy)
```

```python
print("Correct predictions:", np.sum(y_pred == y_test))
print("Total test samples:", len(y_test))
```
